```python
from alpha_vantage.timeseries import TimeSeries
import pandas as pd
import numpy as np


payor=pd.read_csv("D:\payor_stock_code.csv", delimiter=';')
not_duplicate_code=payor.drop_duplicates(['bmbl_stock_code'], keep='last')
list_code=not_duplicate_code['bmbl_stock_code'].tolist()
print(list_code)


exist=[]
notexist=[]
```

```python
for nama in list_code:
    try:
        print('On Progress pull the stock data of '+nama)
        ts = TimeSeries(key='6ODHN889RQZVJVA4', output_format='pandas')
        data, meta_data = ts.get_daily(symbol=nama, outputsize='full')
        data.reset_index(level=0, inplace=True)
        data.to_csv("D:\Data API Alpha Vantage\\1"+nama+".csv", index=None, header=True)
    except:
        print('The stock data of '+nama+' DOES NOT EXIST')
        notexist.append(nama)
    else:
        print('The stock data of '+nama+' SUCCESSFULLY pulled and saved')
        exist.append(nama)
print("Exist :")
print(exist)
print("Not exist :")
print(notexist)
```
